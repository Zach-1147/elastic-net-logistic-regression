```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from chembl_webresource_client.new_client import new_client
from rdkit import Chem
from rdkit.Chem import Descriptors,Lipinski,rdchem
```

```python
target_query = new_client.target.search("Erbb1")
targets = pd.DataFrame(target_query)
targets

selected_query = targets.target_chembl_id[0]
selected_query

activity = new_client.activity
erbb1_ic50 = activity.filter(target_chembl_id = selected_query).filter(standard_type = "IC50")
erbb1_df = pd.DataFrame(erbb1_ic50)

erbb1_df_naremove = erbb1_df[erbb1_df.standard_value.notna()]
erbb1_df_naremove.shape
erbb1_df_naremove.head()

erbb1_df_naremove.columns

selected_columns = ['canonical_smiles', 'molecule_chembl_id','bao_label', 'standard_units', 'standard_value']
erbb1_df_select_columns = erbb1_df_naremove[selected_columns]
erbb1_df_select_columns.head()

erbb1_df_select_columns['bao_label'].unique()
erbb1_df_select_columns['bao_label'].value_counts()

#standardize these
erbb1_df_select_columns['standard_units'].unique()

#look at the range for IC50 values - log transformation?
column_range = int(erbb1_df_select_columns['standard_value']).max() - int(erbb1_df_select_columns['standard_value']).min()
column_range

#Selecting 'single protein format' boa labels
erbb1_spf_df = erbb1_df_select_columns[erbb1_df_select_columns['bao_label'] == 'single protein format']
erbb1_spf_df.shape

#Removing duplicate compounds 
erbb1_spf_noduplicates_df = erbb1_spf_df.drop_duplicates(subset=['molecule_chembl_id'])
erbb1_spf_noduplicates_df.shape
```
